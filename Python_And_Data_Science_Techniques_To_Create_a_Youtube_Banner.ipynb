```python
mental_health_keywords = ["Fight-stress","Defeat-depression", "Fight-anxiety", "Attract-Love","Love-therapy", "Believe-counseling", "Improve mental health", "Positive-Attitude","SelfLove"]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

keywords_text = " ".join(mental_health_keywords)

width = 2560  # Recommended width for YouTube banner
height = 423  # Minimum height for YouTube banner

# Create the word cloud object with the adjusted dimensions
wordcloud = WordCloud(
   background_color="white",
   width=width,
   height=height,
   colormap="viridis"
).generate(keywords_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(16, 2.65))  # Adjust the figure size for the new dimensions
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
